```python
from tensorflow.python.keras.datasets import imdb

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words = 10000)
```

```python
print(x_train[0])
```

```python
print(y_train[0])
```

```python
class_names = ['Negative', 'Positive']
```

```python
word_index = imdb.get_word_index()
print(word_index['brilliant'])
```

# Reverse Word Index

```python
reverse_word_index = dict((value, key) for key, value in word_index.items())
```

#### dict is not callable, so creating a new function decode to return the reversed-word

```python
def decode(sample):
    text = ' '
    for i in sample:
        text += reverse_word_index[i]
        text += ' '
    return text
```

```python
decode(x_train[0])
```

```python
def show_length():
    print('Length of 1st training sample =', len(x_train[0]))
    print('Length of 2nd training sample =', len(x_train[1]))
    print('Length of 1st test sample =', len(x_test[0]))
    print('Length of 2nd test sample =', len(x_test[1]))
    
show_length()
```

# Padding the sample:

-The sample are padded with 'the' since it is just an article and doesn't have any specific meaning to it.

-Post padding method is used.

```python
word_index['the']
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
x_train = pad_sequences(x_train, value = word_index['the'], padding = 'post', maxlen = 256)
x_test = pad_sequences(x_test, value = word_index['the'], padding = 'post', maxlen = 256)
show_length()
```

```python
decode(x_train[0])
```

# Word Embeddings:

### embeddings are done to find correlation between words, so that the model can extrapolate and learn.

## Training the model:

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Embedding, GlobalAveragePooling1D, Dropout

model = Sequential([
    Embedding(10000, 16),
    GlobalAveragePooling1D(),
    Dense(16, activation = 'relu'),
    Dense(1, activation = 'sigmoid')
])
# model.add(Dropout(0.1))

model.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

model.summary()
```

```python
from tensorflow.python.keras.callbacks import LambdaCallback
simple_log = LambdaCallback(on_epoch_end = lambda e, l: print(e, end = '.'))

E = 20

h = model.fit(
    x_train, y_train,
    validation_split = 0.2,
    epochs = E,
    callbacks = [simple_log],
    verbose = 1
)
```

### Predictions and evaluation:

```python
loss ,acc = model.evaluate(x_test, y_test)
print("Test accuracy:", acc *100)
```
